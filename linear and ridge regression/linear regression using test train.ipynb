```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\Shivam\Downloads\carprices.csv")
data.sample(5)
```

```python
dummies=pd.get_dummies(data.carmodel)
```

```python
newData=pd.concat([data,dummies],axis='columns')
```

```python
finalData=newData.drop(['carmodel','Mercedez Benz C class',],axis='columns')
```

```python
model=LinearRegression()
model
```

```python
X=finalData.drop(['sellingprice'],axis='columns')
```

```python
Y=finalData.sellingprice
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=11)
                                            
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
Y_test
```

```python
model.score(X_test,Y_test)
```
