```python
import pandas as py
import numpy as np
from sklearn.linear_model import Ridge
```

```python
data=py.read_csv(r"C:\Users\Shivam\Downloads\carprices.csv")
```

```python
dummie=py.get_dummies(data.carmodel)
```

```python
n=py.concat([dummie,data],axis='columns')

            
```

```python
newData=n.drop(['Mercedez Benz C class','carmodel'],axis='columns')
```

```python
x=newData.drop(['sellingprice'],axis='columns')
x
```

```python
y=newData.sellingprice
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
model=Ridge( max_iter=10)
model
```

```python
valuesAlpha=[]
for i in  range (1,101):
    valuesAlpha.append(i)

print(valuesAlpha)
print(type(valuesAlpha))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

ridgecv=GridSearchCV(
    estimator = model,
    param_grid = {'alpha':valuesAlpha},
    scoring ='neg_mean_squared_error',
    cv =5,
    verbose=0
)
ridgecv.fit(x_train,y_train)
print(ridgecv.best_params_)

                         
           
                     
                 
```

```python
ridgecv.predict(x_test)
```

```python
ridgecv.score(x_test,y_test)
```

```python
ridgecv.best_score_
```

```python
df=py.DataFrame(ridgecv.cv_results_)
df
```
